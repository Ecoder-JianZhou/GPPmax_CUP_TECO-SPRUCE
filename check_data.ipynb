```python
import pandas as pd
import matplotlib.pyplot as plt

iplot = "P20"
isps  = "Shrub"
year  = 2010

weight = 0.5 if isps == "Tree" else 0.25

df_new = pd.read_csv(f"../TECO_outputs/run_mcmc_alltreat_{iplot}/results_csv_format/TECO-SPRUCE_run_mcmc_alltreat_{iplot}_Daily.csv")
df_old = pd.read_csv(f"../outputs_mod/TECO-SPRUCE_run_mcmc_alltreat_{iplot}_Daily.csv")

df_new = df_new[df_new[" year"]>year].copy()
df_old = df_old[df_old[" year"]>year].copy()

fig, axes = plt.subplots(1, 2, figsize=(12,6))

# plot
axes[0].scatter(df_old.index, df_old[f"gpp_{isps}"]*weight*24)
axes[0].scatter(df_new.index, df_new[f"gpp_{isps}"]*weight*24)
import matplotlib.dates as mdates

ax = axes[0]

ax.xaxis.set_major_locator(mdates.YearLocator())   
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# 1:1 
axes[1].scatter(df_old[f"gpp_{isps}"]*weight*24, df_new[f"gpp_{isps}"]*weight*24)

```
